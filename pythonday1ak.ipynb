```python
#accept project marks, internal marks, external marks from student
#passmarks is 50 in each subject
#calculate total score if student got more than or equal to 50 in each subject.
#total score=70% on project+20% on extrenal +10% on internal
#print A grade if score is 90+, B if score is b/w 70-90, C if score is b/w 50-70

project=int(input("Enter project marks"))     #50
internal=int(input("Enter internal marks"))   # 50
external=int(input("Enter external marks"))   # 50
if project>=50 and internal>=50 and external>=50:
    total=(project*70)/100+ (external*20)/100+(internal*10)/100
    print("Total score is: ",total)
    if total>=90:
        print("A grade")
    elif total>=70:
        print("B grade")
    else:
        print("C grade")

```

```python
#accept start and stop value and print b/w values
start=int(input("enter your start value:"))
stop=int(input("enter your stop value"))
if start<stop:
   for i in range(start,stop):
     print(i,end=" ")
elif start>stop:
   for i in range(stop,start,-1):
    print(i,end=" ")
else:
  print("your start and stop values are same")
```
